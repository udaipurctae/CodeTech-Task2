## ORDINAL ENCODING by Scikit learn and by math

```python
df1=pd.DataFrame({"Size":["s","m","l","xl","s","xxl","l","xl","s","m","l","xxl"]})
```

```python
df1.head(2)
```

```python
ord_data=[["s","m","l","xl","xxl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df1[["Size"]])
```

```python
df1["Size_en"]= oe.transform(df1[["Size"]])
```

```python
ord_data1 = {"s":10,"m":11,"l":12,"xl":13,"xxl":14}
```

```python
df1["Size_en_map"] = df1["Size"].map(ord_data1)
df1
```

#### IQR stands for Inter-Quartile Range

```python
# Q1 is the 25% of original data and 3rd quartile contains 75% data
# IQR is obtained by subtracting Q3 and Q1
```

```python
dataset.shape
```

```python
q1=dataset["vehicle_age"].quantile(0.25)
q3 = dataset["vehicle_age"].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
min_range = q1 - (1.5*iqr)
max_range = q3 + (1.5*iqr)
min_range,max_range
```

```python
new_dataset=dataset[dataset["vehicle_age"]<=max_range]
```

```python
new_dataset.shape
```

```python
plt.figure()
sns.boxplot(x="vehicle_age",data=new_dataset)
plt.show()
```

##  Using   Z   scores   to   remove    outliers  

```python
sns.distplot(dataset["region_density"])
plt.show()
```

```python
dataset.shape
```

```python
min_range = dataset["region_density"].mean() - (3*dataset["region_density"].std())
max_range = dataset["region_density"].mean() + (3*dataset["region_density"].std())
```

```python
min_range,max_range
```

```python
newdata=dataset[dataset["region_density"]<=max_range]
```

```python
sns.boxplot(x="region_density",data=newdata)
plt.show()
```

```python
dataset.shape
```

```python
z_score=(dataset["region_density"] - (dataset["region_density"].mean())/dataset["region_density"].std())
```

```python
dataset["z_score"] = z_score
```

```python
dataset[dataset["z_score"]<3*z_score]
```
